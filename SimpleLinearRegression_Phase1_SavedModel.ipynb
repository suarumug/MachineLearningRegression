```python
# Importing pandas,all class and its functions here to read the dataset
import pandas as pd # providing alias name to call in short rather pandas everytime 
```

```python
# Using pandas built in function "read_csv", reading csv file which is dataset for this prediction 
# and saving all data in variable called"dataset"
dataset=pd.read_csv("Salary_Data.CSV")
```

```python
# Vieving the data by executing the variable name. 
dataset
```

```python
# Splitting the input and output columns. 
# Assiging the input column and output column to separate variable from dataset variable.
# input columns values are independent  and output column values are dependent
# At below, assigning input column to variable called"independent" and output column name to "dependent" variable 
# we can assign multile columns to single variale byt providing comma (,) separator.
independent=dataset[["YearsExperience"]]
dependent=dataset[["Salary"]]
```

```python
# executing input column assigned variabel to view the values
independent
```

```python
## executing input column assigned variabel to view the values
dependent
```

```python
# separating the dataset to training set and test set 
# for that, we need to import Sklearn libraries 
from sklearn.model_selection import train_test_split
# splitting the dataset to trainign set and test set and saving the set to the variable
# 70% of input data assigned to input training set "x_train" and 30 % assigned to input test set "x_test"
# 70% of output data assigned to output training set "y_train" and 30 % of output dataset assigned to output test set "y_test"
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
# Model creation phase
# Importing linear regression from sklearn libraries
from sklearn.linear_model import LinearRegression
#assigning LinearRegression class to a non parametrized variable so all the functions in LinearRegression class is saved to regressor variable. 
#In other words, regressor variable can able to access all teh functions in LinearRegression class.
regressor=LinearRegression()
#calling the function "fit" by passing the input and output values of training set to the function, so that algoritham learns and creates the model
regressor.fit(x_train,y_train)

```

```python
# to view the weight value in equation, y=wx+b
weight = regressor.coef_
```

```python
weight
```

```python
# To view the bias values in equation, y=wx+b
bias = regressor.intercept_
```

```python
bias
```

```python
# In above steps,identified teh weight and bias, so model is ready.
#To test the model, which  is, to get the predicted output using test input data, need to use predict function 
y_predict=regressor.predict(x_test)
```

```python
y_predict
```

```python
#Evaluation Metrics:R2 evaluation - CLose to 1 is better model performance and close to 0 is poor model performance
# import the required libraries for R2 evaluation 
from sklearn.metrics import r2_score
# call r2_score function by passing the test output value from dataset and predicted output value and save the result in variable
r_score=r2_score(y_test,y_predict)
```

```python
r_score
```

```python
# Saving model 
# to save, import pickle libraries and craete a filename to save the craeted model in the name of the filename
# Filename should always be saved in .sav extension. Pickle filename should always in .sav extension 
import pickle
filename="finalized_model_linear.sav"


```

```python
# In above steps, filename is just created but model is not saved in the filename yet
# next, to save the model to the filename, 
pickle.dump(regressor,open(filename,"wb"))
```

```python
# loading the saved model to make it ready for deployment
# can be given the actual filename or the variable name of the filename 
#loaded_model=pickle.load(open("finalized_model_linear.sav","rb"))
loaded_model=pickle.load(open(filename,"rb"))
```
